```python
#Location with maximum rainfall
df.groupby('Location')['Rainfall'].sum().idxmax()
```

```python
#split the date into year, month and day
df['Year']=df['Date'].str.split('-').str[0]
df['Month']=df['Date'].str.split('-').str[1]
df['Day']=df['Date'].str.split('-').str[2]
df.head()
```

```python
#drop the date column
df.drop('Date',axis=1,inplace=True)
df.head()
```

```python
#convert the datatypes into int
df['Year']=df['Year'].astype(int)
df['Month']=df['Month'].astype(int)
df['Day']=df['Day'].astype(int)
df[['Year','Month','Day']].dtypes
```

Part 2 – EDA (Exploratory Data Analysis)

```python
#Histogram of rainfall amounts
plt.hist(df['Rainfall'],bins=50,color='green')
plt.xlabel('Rainfall')
plt.ylabel('Frequency')
plt.title('Rainfall Distribution')
plt.savefig('plots/rainfall_distribution.png')
plt.show()
```

```python
#Top 10 location with highest total rainfall
max_rainfall=df.groupby('Location')['Rainfall'].sum()
max_rainfall.sort_values(ascending=False).head(10)
```

```python
#Scatter plot: Humidity vs rainfall
fig,axs=plt.subplots(1,2,figsize=(15,5))

axs[0].scatter(df['Humidity3pm'],df['Rainfall'],color='red')
axs[0].set_xlabel('Humidity3pm')
axs[0].set_ylabel('Rainfall')
axs[0].set_title('Humidity at 3pm vs Rainfall')

axs[1].scatter(df['Humidity9am'],df['Rainfall'],color='blue',label='Humidity9am')
axs[1].set_xlabel('Humidity9am')
axs[1].set_ylabel('Rainfall')
axs[1].set_title('Humidity at 9am vs Rainfall')

plt.savefig('plots/humidity_vs_rainfall.png')
plt.show()
```

```python
#Barchart showing total rainfall per location
plt.figure(figsize=(15,10))
total_rainfall_per_location=df.groupby('Location')['Rainfall'].sum().sort_values(ascending=False)
plt.bar(total_rainfall_per_location.index,total_rainfall_per_location.values,color='brown')
plt.xlabel('Location')
plt.ylabel('Rainfall')
plt.title('Total Rainfall per Location')
plt.xticks(rotation=90)
plt.tight_layout()        #ensures labels don't get cut off
plt.savefig('plots/rainfall_per_location.png')
plt.show()
```

```python
#boxplot of overall humidity distribution
fig,axs=plt.subplots(1,2,figsize=(15,5))

axs[0].boxplot(df['Humidity3pm'],patch_artist=True)
axs[0].set_ylabel('Humidity (%)')
axs[0].set_title('Box plot of humidity at 3pm')

axs[1].boxplot(df['Humidity9am'],patch_artist=True)
axs[1].set_ylabel('Humidity (%)')
axs[1].set_title('Box plot of humidity at 9am')

plt.savefig('plots/overall_humidity.png')
plt.show()
```

```python
#yearly rainfall trend
yearly_rainfall=df.groupby('Year')['Rainfall'].sum()
plt.plot(yearly_rainfall.index,yearly_rainfall.values,marker='o',color='darkblue')
plt.xlabel('Year')
plt.ylabel('Rainfall')
plt.title('Yearly Rainfall Trend')
plt.grid()
plt.savefig('plots/yearly_rainfall_trend.png')
plt.show()
```

```python
#Count of days with/without wind change using barplot
df['wind_change']=(df['WindDir9am']!=df['WindDir3pm']).astype(int)
wind_change_count=df['wind_change'].value_counts()
plt.bar(['No change','Change'],wind_change_count.values,color=['red','darkblue'])
plt.xlabel('Wind change status')
plt.ylabel('Number of days')
plt.title('Distribution of wind direction changes')
plt.savefig('plots/wind_direction.png')
plt.show()
```

Part 3 – Statistics

```python
#mean, median, max rainfall
print('Average Rainfall: ',df['Rainfall'].mean())
print('Median Rainfall: ',df['Rainfall'].median())
print('Maximum Rainfall: ',df['Rainfall'].max())
```

```python
#variance, std deviation of evaporation
print('Variance: ',df['Evaporation'].var())
print('Standard devation: ',df['Evaporation'].std())
```

```python
numerator=(df['RainTomorrow']=='Yes').sum()
denominator=df.shape[0]
probability=numerator/denominator
probability
```

```python
#if it rains today, what is the probability that it will also rain tomorrow
#hint: conditional probability
#formula: P(rain tomorrow| rain today)= P(rain today) intersection P(rain tomorrow) / P(rain today)
numerator=((df['RainToday']=='Yes') & (df['RainTomorrow']=='Yes')).sum()
denominator=(df['RainToday']=='Yes').sum()
probability=numerator/denominator
probability
```

```python
#Percentage of overall average sunshine on rainy days
numerator=df[df['Rainfall']>0]['Sunshine'].mean()
denominator=df['Sunshine'].mean()
percentage=(numerator/denominator)*100
percentage
```

Part 4 – Linear Algebra & NumPy

```python
#create NumPy arrays for monthly Rainfall (in mm) and corresponding Evaporation levels
monthly_rainfall=df.groupby('Month')['Rainfall'].sum()
monthly_evaporation=df.groupby("Month")['Evaporation'].sum()
month_rainfall_array=np.array(monthly_rainfall)
monthly_evaporation_array=np.array(monthly_evaporation)
print('Monthly rainfall array: ',month_rainfall_array)
print('Monthly evaporation array: ',monthly_evaporation_array)
```

```python
#perform addition and subtraction
add=np.add(month_rainfall_array,monthly_evaporation_array)
sub=np.subtract(month_rainfall_array,monthly_evaporation_array)
print('Addition of two arrays: ',add)
print('Subtraction of two vectors: ',sub)
```

```python
#perform dot product between these arrays
dot_product=np.dot(month_rainfall_array,monthly_evaporation_array)
dot_product
```

```python
#normalize the min temp array
min=df['MinTemp'].min()
max=df['MinTemp'].max()
normalized_array=(df['MinTemp']-min)/(max-min)
normalized_array.values
```

Part 5 – Calculus

```python
#diff ELI=(evaporation-rainfall)/evaporation w.r.t evaporation
import sympy as sp
evaporation,rainfall=sp.symbols('evaporation rainfall')
eli=(evaporation-rainfall)/evaporation
diff=sp.diff(eli,evaporation)
diff
```

Part 6 – Feature Engineering

```python
df['is_heavy_rain']=df['Rainfall'].apply(lambda x: 1 if x>50 else 0)
df['is_hot_day']=df['MaxTemp'].apply(lambda x: 1 if x>35 else 0)
df.head()
```

```python
df['temp_range']=df['MaxTemp']-df['MinTemp']
df.head()
```

```python
df['pressure_trend']=df['Pressure9am']-df['Pressure3pm']
df.head()
```

```python
#rank --> ensures uniqueness
#This prevents qcut from failing when many rows have the same value (like 0.0) because qcut needs unique values to create bins.
df['Rainfall_quartiles']=pd.qcut(df['Rainfall'].rank(method='first'),q=4,labels=['low','medium','high','very high'],duplicates='drop')
df.head(10)
```

```python
df['heat_index']=(df['MaxTemp']*df['Humidity3pm'])/100
df.head()
```

```python
#annot --> to show numbers inside each cell
#cmap --> color map (controls the color gradient of the heatmap)
#fmt --> format (how to display numbers when annot=True)
      # "d" --> int, ".1f" --> one decimal place, ".2f" --> two decimals
#cbar --> color bar (the gradient bar at the side of the heatmap)

columns=['Rainfall','Sunshine','MaxTemp','MinTemp','Humidity9am','Humidity3pm','WindSpeed9am','WindSpeed3pm']
correlation=df[columns].corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation,annot=True,cmap='coolwarm',fmt='.2f',cbar=True)
plt.title('Correlation Heatmap of Weather Variables')
plt.savefig('plots/correlation_heatmap.png')
plt.show()
```
